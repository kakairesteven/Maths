<h4>Functions</h4>
<p>Expressions that define relationships between two or variables</p>
<p>f(x) = 4*x + 6</p>

```python
# Declaring a linear function
def f(x):
    return 2*x + 1

x_values = [2, 5, 3, 9]
for x in x_values: 
    y = f(x)
    print(y)
```

```python
from sympy import *
x = symbols('x')
f = 2*x + 1
plot(f)
```

<h3>Summation</h3>

```python
summation = sum(2*i for i in range(1, 6))
summation
```

```python
x = [1, 4, 6, 2]
n = len(x)
summation = sum(10*x[i] for i in range(0, n))
summation
```

```python
i, n = symbols('i n')
summation = Sum(2*i, (i, 1, n))
up_to_5 = summation.subs(n ,5)
up_to_5
```

```python
up_to_5.doit()
```
