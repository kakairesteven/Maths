<h3>Derivatives</h3>

```python
def derivative_x(f, x, step_size):
    m = (f(x + step_size) - f(x)) / ((x + step_size) - x)
    return m
```

```python
def my_function(x):
    return x**2
```

```python
derivative_x(my_function, 2, .000001)
```

```python
# calculating derivative using limit
# Declaring varibles.
x, s = symbols('x s')

#Function, f(x)
f = x**2
slope_f = (f.subs(x, x+s)-f) / ((x+s)-x)

# Slope/gradient at x=2
slope_2 = slope_f.subs(x, 2)

# value as s tends to zero but never reaches zero
result = limit(slope_2, s, 0)
result
```

```python
#The chain rule has applications in designing neural networks,
# We shall get to that later.
#Examples 
y = x**2 + 1
z = y**3 - 2

#Subsituting for y in z
z = (x**2 + 1)**3 - 2
dz_dx = diff(z)
dz_dx
```

```python
#using the chain rule
#using different variable to avoid the earlier defined variables from
#interfering with this example.
#Define symbols used
a, b = symbols('a b')

#functions
_b = a**2 + 1
c = b**3 - 2

# Derivatives, db_da and dc_da
db_da = diff(_b)
dc_db = diff(c)

#the chain rule with substitution
dc_da = (db_da * dc_db).subs(b, _b)
dc_da
```

<h3>Integral</h3>

```python
#integration is the opposite of differentiation.
#integral is the summation of infinitesmall elements from a given point x1 to a point x2.
#integration finds the area under a curve.
```

```python
# approximating an integral by Riemann sum/the rectangule rule
def approximate_integral(a, b, n, f):
    delta_x = (b-a)/n
    total_sum = 0

    for i in range(1, n+1):
        midpoint = 0.5*(2*a + delta_x*(2*i-1))
        total_sum += f(midpoint)
    return total_sum*delta_x

def my_function(x):
    return x**2 + 1

area = approximate_integral(a=0, b=1, n=1_000_000, f=my_function)
area
```

```python
#using sympy
x = symbols('x')
f = x**2+1
area = integrate(f, (x, 0, 1))
area
```

```python
#using limits to calculate integral
#define symbols
x, i, n = symbols('x i n')

#lower and upper limit
lower, upper = 0, 1

#small change, dx
delta_x = (lower + upper)/n

#x value at a position i, x_i
x_i = lower + delta_x*i

#y_value at a position x_i
fx_i = f.subs(x, x_i)

#using the summation function to sum all rectangles form i=1 to i=n
n_rectangles = Sum(delta_x*fx_i, (i, 1, n)).doit()

#using limit function to approximate area, as n tends to zero
area = limit(n_rectangles, n, oo)
area
```
